**Importing The Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

 **Importing The Dataset**

```python
traindataset=pd.read_csv('train_data.csv')
x= traindataset.iloc[:,2:6].values
y=traindataset.iloc[:,-1].values
testdataset=pd.read_csv('unique.csv')
x_unique=testdataset.iloc[:,2:]
```

**Splitting  The Dataset**

```python
# from sklearn.model_selection import train_test_split
# x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

**Training The Random Forest Model**

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=1000,random_state=0)
regressor.fit(x,y)
```

**Predicting The Values**

```python
y_pred=[]
for i in x:
  l=[i]
  sample=regressor.predict(l)
  y_pred.append(sample)
y_pred
```

**Displaying Actual Values VS Predicted Values**

```python
y_pred_df = pd.DataFrame({'Actual Value':y,'Predicted Values': y_pred})
y_pred_df
```

**Checking The Accuracy**

```python
from sklearn.metrics import r2_score
print('Accuracy: %.2f'% r2_score(y,y_pred))
```

**Finding The Delay Time Of The New Data**

```python
y_unique= regressor.predict(x_unique)
print(y_unique)
```

**Displaying the Predicted Values of the Unique Data**

```python
y_unique_df = pd.DataFrame({'Predicted Values': y_unique})
y_unique_df
```

**Saving the model**

```python
import pickle
filename= 'train_model.pkl'
pickle.dump(regressor,open(filename,'wb'))
```

**Loading Model**

```python
loaded_model=pickle.load(open(filename,'rb'))
y_loaded=loaded_model.predict(x_unique)
print(y_loaded)
```
